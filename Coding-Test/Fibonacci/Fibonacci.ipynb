```python
def fib(n):
  _curr, _next = 0, 1
  for _ in range(n):
    _curr, _next = _next, _curr + _next
  return _curr
```

```python
fib(3)
```

```python
def fib(n):
  _curr, _next = 0, 1
  for _ in range(n+1):
    yield _curr
    _curr, _next = _next, _curr + _next
```

```python
fib=fib(5)
for i, val in enumerate(fib):
  print('Fibonacci({}) : {}'.format(i, str(val)))
```

```python
def fib(n):
  if n==0:
    return 0
  elif n==1 or n==2:
    return 1
  else:
    return fib(n-1) + fib(n-2)
```

```python
fib(8)
```

```python
from functools import lru_cache

@lru_cache(maxsize=None)
def fib(n):
  if n==0:
    return 0
  elif n==1 or n==2:
    return 1
  else:
    return fib(n-1) + fib(n-2)
```

```python
n=50
print('Fibonacci({}) : {}'.format(n, fib(n)))
```
